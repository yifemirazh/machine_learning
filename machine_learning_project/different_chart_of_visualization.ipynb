```python
import matplotlib.pyplot as plt

# Data
countries = ['USA', 'India', 'China']
population = [330, 1400, 1440]

# Plot
plt.bar(countries, population, color='skyblue')
plt.title('Population by Country')
plt.xlabel('Country')
plt.ylabel('Population (in Millions)')
plt.show()
```

bar chart to show population numbers in millions by different countries category

```python
# Data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [100, 120, 130, 90, 150]

# Plot
plt.plot(months, sales, marker='o', color='green')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

line chart to show sales by usinng months category

```python
# Data
labels = ['Apples', 'Bananas', 'Cherries']
sizes = [40, 35, 25]

# Plot
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Fruit Market Share')
plt.axis('equal')  # Equal aspect ratio ensures the pie is round.
plt.show()
```

graphs of pie chart to show fruit market share

```python
import seaborn as sns

# Data
height = [150, 160, 170, 180, 190]
weight = [50, 60, 65, 80, 90]

# Plot
sns.scatterplot(x=height, y=weight)
plt.title('Height vs Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
```

scatetr plot to show the relation ship between weight and height

```python
# Data
scores = [50, 55, 60, 60, 65, 70, 75, 80, 85, 90, 95, 100]

# Plot
plt.hist(scores, bins=5, color='orange', edgecolor='red')
plt.title('Score Distribution')
plt.xlabel('Score')
plt.ylabel('Number of Students')
plt.show()
```

histogram to show the distribution of scores by student number

```python
import numpy as np

# Random data
data = np.random.rand(5,5)

# Plot
sns.heatmap(data, annot=True, cmap='coolwarm')
plt.title('Heat Map Example')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [10, 20, 30, 40, 50]       # x-axis
y = [15, 25, 35, 45, 55]       # y-axis
bubble_size = [100, 300, 500, 700, 900]  # Size of bubbles

# Plot
plt.scatter(x, y, s=bubble_size, alpha=0.5, c="purple")
plt.title('Bubble Chart Example')
plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.grid(True)
plt.show()
```

```python
import squarify
import matplotlib.pyplot as plt

# Data
values = [500, 300, 200, 100]
labels = ["USA", "India", "China", "Germany"]

# Plot
squarify.plot(sizes=values, label=labels, color=["red", "blue", "green", "orange"], alpha=0.7)
plt.title("Tree Map Example - Country Size")
plt.axis('off')  # Hide axes
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load world map data
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Plot the world map
world.plot(figsize=(12, 8), color='lightblue', edgecolor='black')
plt.title('Simple World Map')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example data
data = {'Scores': [3, 5, 7, 8, 9, 11, 13, 15, 16]}
df = pd.DataFrame(data)

# Create a box plot
sns.boxplot(x=df['Scores'])
plt.show()

# Calculate Quartiles
Q1 = df['Scores'].quantile(0.25)
Q2 = df['Scores'].median()
Q3 = df['Scores'].quantile(0.75)
IQR = Q3 - Q1

print("Q1:", Q1)
print("Median (Q2):", Q2)
print("Q3:", Q3)
print("IQR:", IQR)
```
