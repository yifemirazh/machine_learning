```python
import matplotlib.pyplot as plt
import numpy as np
weeks = list(range(1,11))
group_a = [3.1, 3.0, 2.9, 3.2, 3.1, 2.8, 3.0, 3.1, 2.9, 3.0]
group_b = [2.8, 2.6, 2.7, 2.9, 2.5, 2.6, 3.0, 2.8, 2.7, 2.9]
avg_time = [(a + b) / 2 for a, b in zip(group_a, group_b)]
# Calculate the average time spent in class by students in two groups over a period of 10 weeks
mean=np.mean(avg_time)
std_dev=np.std(avg_time)
ucl=mean + 3 * std_dev

lcl=mean - 3 * std_dev
plt.figure
plt.plot(weeks, avg_time, marker="o", color="purple", label='average time(all students)')
plt.axhline(mean, color='green', linestyle='--', label='mean')
plt.axhline(y=ucl, color='red', linestyle='--', label='upper control limit (UCL)')
plt.axhline(y=lcl, color='blue', linestyle='--', label='lower control limit (LCL)')
plt.title('x_bar chart :weelky average class time (combined groups) with conrol limits')
plt.xlabel('Weeks')
plt.ylabel('Average time (hours)')
plt.xticks(weeks)
# Example: Set ticks at y=1, y=2.5, and y=4
plt.yticks([1, 2.5, 4])
plt.grid(True) 
plt.legend()
plt.show()  
#This code is used to plot the average time spent in class by students in two groups over a period of 10 weeks.

```

```python
import matplotlib.pyplot as plt
import numpy as np

late_students=[3,4,5,2,3,4,6,2,3,1,4,5,3,4,2,3,5,4,6,3]
n=37
proportions_late=[x/n for x in late_students]
p_bar=np.mean(proportions_late)
std_dev=np.std(proportions_late)
ucl=p_bar + 3 * std_dev 
lcl=p_bar - 3 * std_dev 
plt.figure(figsize=(10, 6))  
plt.plot(range(1,21), proportions_late, marker="o", color="purple", label='proportions_late students')   
plt.axhline(y=p_bar, color='black', linestyle='--', label=f'average proportion of late students ({p_bar:.3f})') 
plt.axhline(y=ucl, color='red', linestyle='--', label=f'upper control limit (UCL)={ucl:.3f}')  
plt.axhline(y=lcl, color='green', linestyle='--', label=f'lower control limit (LCL)={lcl:.3f}') 
plt.title('p chart: weekly proportion of late students with control limits')    
plt.xlabel('days') 
plt.ylabel('Proportion of late students')   
plt.xticks(range(1, 21))   
plt.ylim(-0.011, 0.25)
plt.axhline(y=0, color='black', linewidth=1)
plt.grid(True)
plt.legend() 
plt.tight_layout()   
plt.show()            
```
